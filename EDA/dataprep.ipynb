```python
import pandas as pd
import numpy as np

excel = pd.ExcelFile("../data/res2_1-5.xlsx")

print(excel.sheet_names)
```

```python
res1 = excel.parse('res1')
res1['level'] = 'beginner'

res2 = excel.parse('res2')
res2['level'] = 'intermediate'

res3 = excel.parse('res3')
res3['level'] = 'upper_intermediate'

res4 = excel.parse('res4')
res4['level'] = 'advanced'

res5 = excel.parse('res5')
res5['level'] = 'academic'

combined_data = pd.concat([res1, res2, res3, res4, res5])

combined_data
```

```python
combined_data['AvgWord in Syl'] = combined_data.groupby('level')['AvgWord in Syl'].transform(
    lambda x: x.fillna(x.mean())
)

combined_data['AvgWord in Syl'].fillna(combined_data['AvgWord in Syl'].mean(), inplace=True)
combined_data
```

```python
cols_to_keep = [
    'Words', 'Syllables', 'Sentences', 'Letters',
    'AvgWord in Letters', 'AvgWord in Syl', 'AvgSent',
    'level'
]

cleaned_data = combined_data[cols_to_keep].copy()

print(cleaned_data.head())
```

```python
cleaned_data.rename(columns={
    'Words': 'words',
    'Syllables': 'syllables',
    'Sentences': 'sentences',
    'Letters': 'letters',
    'AvgWord in Letters': 'avg_word_len',
    'AvgWord in Syl': 'avg_word_syl',
    'AvgSent': 'avg_sent_len'
}, inplace=True)

cleaned_data
```

```python
cleaned_data = cleaned_data.round(3)
cleaned_data.to_csv("../data/cleaned_data.csv", index=False)
print("Cleaned dataset saved, Master!")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = cleaned_data[['words', 'sentences', 'letters', 'avg_word_len', 'avg_sent_len', 'syllables', 'avg_word_syl']]
y = cleaned_data['level']

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
#from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler()
#X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
perm_imp = pd.DataFrame({'Feature': X.columns, 'Importance': result.importances_mean})
perm_imp = perm_imp.sort_values('Importance', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(data=perm_imp, x='Importance', y='Feature', palette='viridis')
plt.title("Permutation Feature Importance", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    random_state=42
)
xgb.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=3, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
print("Best score:", grid.best_score_)
```
